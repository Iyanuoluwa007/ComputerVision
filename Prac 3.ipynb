```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blk_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blk_img.shape
```

```python
plt.imshow(blk_img)
```

```python
cv2.rectangle(blk_img,pt1=(384,10),pt2=(500,150),color=(0,255,0),thickness=10)
plt.imshow(blk_img)
```

```python
cv2.rectangle(blk_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
plt.imshow(blk_img)
```

```python
cv2.circle(img=blk_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blk_img)
```

```python
cv2.circle(img=blk_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blk_img)
```

```python
cv2.circle(img=blk_img, center=(400,100), radius=50, color=(255,0,0), thickness=-5)
plt.imshow(blk_img)
```

```python
cv2.line(blk_img,pt1=(0,0),pt2=(512,512), color=(102,255,255),thickness=5)
plt.imshow(blk_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
blk_img2 = cv2.putText(blk_img,text='Hello',org=(10,500),fontFace=font, fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blk_img2)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
blk_img3 = cv2.putText(blk_img,text='Hello',org=(10,500),fontFace=font, fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_4)
plt.imshow(blk_img3)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blk_img,text='Hello',org=(10,500),fontFace=font, fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_8)
plt.imshow(blk_img)
```
